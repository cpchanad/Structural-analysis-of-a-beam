```python
import planesections as ps

# Define beam
L = 5
beam = ps.newEulerBeam(L)

# Define node locations and support conditions/fixities
fixed = [1,1,1]
pinned = [1,1,0]
roller = [0,1,0]
beam.setFixity(0, pinned, label = 'A')
beam.setFixity(L, pinned, label = 'B')

# Define loads
Pz = -1  # kN or kN/m
# Add point load (in kN)
#beam.addVerticalLoad(L*0.5, 2*Pz, label = 'F')
# Add uniform dist load (in kN/m)
beam.addDistLoadVertical(0, L, Pz)

# Plot the beam diagram
ps.plotBeamDiagram(beam)
```

```python
# Run the analysis
analysis = ps.OpenSeesAnalyzer2D(beam)
analysis.runAnalysis()

# Plot the SFD and BMD
ps.plotShear(beam, yunit = 'kN')
ps.plotMoment(beam, yunit = 'kNm')
```
